```python
# Gradient Descent vs Newton's Method
# Comparative visualization on f(x) = x^3 - 2x - 5

import numpy as np
import matplotlib.pyplot as plt

# Define function, first derivative, and second derivative
def f(x): return x**3 - 2*x - 5
def df(x): return 3*x**2 - 2
def d2f(x): return 6*x

# Initial guess
x0 = 2.5
lr = 0.05  # learning rate for Gradient Descent

# Iterations
gd_points = [x0]
nm_points = [x0]

x_gd, x_nm = x0, x0

for i in range(10):
    # Gradient Descent update
    x_gd = x_gd - lr * df(x_gd)
    gd_points.append(x_gd)
    
    # Newton's Method update
    x_nm = x_nm - df(x_nm)/d2f(x_nm)
    nm_points.append(x_nm)

# Generate data for visualization
x_vals = np.linspace(-3, 3, 400)
y_vals = f(x_vals)

# Plot setup
plt.figure(figsize=(10,6))
plt.plot(x_vals, y_vals, label='f(x)', color='blue')

# Plot iteration paths
plt.scatter(gd_points, [f(x) for x in gd_points], color='orange', label='Gradient Descent', s=60)
plt.scatter(nm_points, [f(x) for x in nm_points], color='red', label='Newton\'s Method', s=60)

# Axis and formatting
plt.axhline(0, color='gray', linestyle='--')
plt.legend()
plt.title("Gradient Descent vs Newton's Method on f(x) = x³ - 2x - 5")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(alpha=0.2)
plt.show()

print(f"Final GD x ≈ {gd_points[-1]:.6f}")
print(f"Final Newton x ≈ {nm_points[-1]:.6f}")
```

# Gradient Descent vs Newton’s Method — Comparative Visualization

**Objective:**  
To compare the convergence behavior of Gradient Descent and Newton’s Method on the same function \( f(x) = x^3 - 2x - 5 \).

---

### Theoretical Insight

- **Gradient Descent:**
  \[
  x_{n+1} = x_n - \eta f'(x_n)
  \]
  Uses only first derivatives. Requires a learning rate \( \eta \) and converges linearly.

- **Newton’s Method:**
  \[
  x_{n+1} = x_n - \frac{f'(x_n)}{f''(x_n)}
  \]
  Uses curvature information (second derivative). Converges quadratically near the root.

---

### Visualization Goal

Both methods start from the same initial guess but follow different update rules.  
This demonstrates how Newton’s Method uses curvature (Hessian information) to make faster, smarter jumps toward the root/minimum.
