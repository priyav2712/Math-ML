```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): return x**3 - 2*x - 5
def df(x): return 3*x**2 - 2

x = 2.0
values = [x]

for i in range(7):
    x = x - f(x)/df(x)
    values.append(x)

print("Root ‚âà", x)

# Visualization
x_vals = np.linspace(-3, 3, 400)
plt.plot(x_vals, f(x_vals), label='f(x)')
plt.axhline(0, color='gray', linestyle='--')
plt.scatter(values, [f(v) for v in values], color='red', label='Iterations')
plt.legend()
plt.title('Newton‚Äôs Method Iterations')
plt.show()
```

# Newton‚Äôs Method ‚Äî Numerical Root Finding

**Objective:**  
To find a real root of \( f(x) = x^3 - 2x - 5 \) using Newton‚Äôs Method and visualize its convergence.

---

### Mathematical Idea

Newton‚Äôs Method iteratively improves an approximation \( x_n \) to a root by using the tangent line of \( f(x) \):

\[
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
\]

If \( f(x) \) is differentiable and the initial guess \( x_0 \) is sufficiently close to the root, the method converges quadratically.

---

### Implementation Overview
- Function: \( f(x) = x^3 - 2x - 5 \)  
- Derivative: \( f'(x) = 3x^2 - 2 \)  
- Initial Guess: \( x_0 = 2.0 \)  
- Iterations: 7  
- Visualization: Each red point shows successive approximations.


---

### üß© Result

The method converges to:

\[
x \approx 2.0945514815423265
\]

This means \( f(x) = 0 \) at approximately **x = 2.09455**, confirming convergence to the unique real root.
